```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
df = pd.read_csv('meteo_data_30.csv')
# Сохраняем столбцы 'Date' и 'HC_AIR_TEMPERATURE', удаляя остальные столбцы
df = df[['Date', 'HC_AIR_TEMPERATURE']]
df.head(5)
# Преобразование столбца 'Date' в тип datetime
df['Date'] = pd.to_datetime(df['Date'])

# Установка столбца 'Date' в качестве индекса
df.set_index('Date', inplace=True)

# Создание модели ARIMA
model = ARIMA(df['HC_AIR_TEMPERATURE'], order=(1, 1, 1))

# Обучение модели ARIMA
model_fit = model.fit()

# Получение прогноза на следующий час
forecast = model_fit.forecast(steps=3)

# Вывод прогноза
print("Прогноз на следующий час:")
print(forecast)

# Визуализация прогноза
plt.plot(df.index, df['HC_AIR_TEMPERATURE'], label='Исходные данные')
plt.plot(pd.date_range(start=df.index[-1], periods=3, freq='H')[1:], forecast, label='Прогноз')
plt.xlabel('Дата')
plt.ylabel('Температура')
plt.legend()

# Установка ограничения по оси x
plt.xlim(df.index[-2], pd.date_range(start=df.index[-1], periods=3, freq='H')[1])

plt.show()
```
